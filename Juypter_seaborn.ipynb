```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

crash_df = sns.load_dataset('car_crashes')
crash_df
```

```python
sns.histplot(crash_df['not_distracted'], bins=20)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
sns.set_context('paper', font_scale=1.4)
#sns.set_context('talk', font_scale=1.4)
#sns.set_context('poster', font_scale=1.4)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde', fill=True)
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
tips_df = sns.load_dataset('tips')
print(tips_df.to_string())
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues', diag_kind='hist')
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

```python
sns.countplot(x='sex', data=tips_df)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.legend(loc=0)
# Median, IQR(Interquartile Range), whiskers, and outliers
```

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(8, 5))
sns.set_style('whitegrid')  # 'white', 'dark', 'whitegrid', 'darkgrid', 'ticks'
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True, palette='magma')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)

crash_mx = crash_df.corr(numeric_only=True)
crash_mx
```

```python
sns.heatmap(crash_mx, annot=True, cmap='coolwarm')
```

```python
flights = sns.load_dataset('flights')
flights 
```

```python
flights =flights.pivot_table(index='month', columns='year', values='passengers')
flights
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1)
plt.yticks(rotation=0) 
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species')
#iris_g.map(plt.scatter)
iris_g.map_diag(plt.hist)
# iris_g.map_offdiag(plt.scatter)
iris_g.map_lower(plt.scatter)
iris_g.map_upper(sns.kdeplot)
plt.legend()
```
